### Seaborn

**Seaborn is a visualization library that sits on top of matplotlib**

*Seaborn offers enhanced features compared to matplotlib*

More Details: https://seaborn.pydata.org/examples/index.html


### Scatter Plots

**Importing Dependencies**

```
import numpy as np
import pandas as pd
import seaborn as sns
```

### Importing Datasets

```
from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()
```

```
dataset
```

### Loading Dataset into DataFrame

```
df = pd.DataFrame(np.c_[dataset['data'], dataset['target']], columns = np.append(dataset['feature_names'], ['target']))
```

```
df.head()
```

### Scatter Plots using Seaborn

```
sns.scatterplot(x = 'mean area', y = 'mean smoothness', data = df)
plt.title("Area vs Smoothness")
plt.legend(loc = 'upper left')
```

**Plotting graph with respect to target**

```
sns.scatterplot(x = 'mean area', y = 'mean smoothness', hue = 'target',  data = df)
plt.title("Area vs Smoothness")
plt.legend(loc = 'upper right')
```

```
sns.scatterplot(x = 'mean radius', y = 'mean area', hue = 'target',  data = df)
plt.title("Error in Calculating")
plt.legend(loc = 'lower right')
```

### Counter Plots

```
sns.countplot(df['target'])
```

### Pair Plots

Pair Plots are used to find the relationship between given variables

```
columns = df.columns
```

```
columns = columns[0:5]
columns
```

```
plt.figure(figsize = (8,5))
sns.pairplot(data = df, hue = 'target', vars = columns)
```

### Heats Maps

Heats maps are used to plot the correlation between the two or more columns

```
plt.figure(figsize = (20, 10))
sns.heatmap(df.corr())
```

Heat map correlations with annotations

```
plt.figure(figsize = (20, 10))
sns.heatmap(df.corr(), annot = True)
```

### Displots

Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)

KDE is used to plot the Probability Density of a continuous variable. 


```
sns.displot(df['mean radius'], bins = 25, color = 'blue')
```

```
sns.distplot(df['mean radius'], bins = 25, color = 'blue')
```

```
class_0_df = df[ df['target'] ==0 ]
class_1_df = df[ df['target'] ==1 ]
```

```
class_0_df
```

```
class_1_df
```

**Plot the distplot for both classes**

```
plt.figure(figsize=(10, 7))
sns.distplot(class_0_df['mean radius'], bins = 25, color = 'blue')
sns.distplot(class_1_df['mean radius'], bins = 25, color = 'red')
plt.grid()
```
